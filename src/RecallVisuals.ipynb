```python
#Toyota
x =[0,1,2]
posy = [306.38,278.58,301.35]
negy= [45.40,92.04,40.35]
comy = [594.85,452.94,606.51]
plt.figure()
plt.plot(x, posy,linewidth=5, color='black', label='Positive')  # on top
plt.plot(x, negy, linewidth=5, color='red', label='Negative')  # bottom
plt.plot(x, comy, linewidth=5, color='green', label='Compound')
#plt.axhline(0, linewidth=10, color='blue', zorder=2)
plt.title('Toyota Sentiment')
l = plt.legend()
l.set_zorder(20)  # put the legend on top
plt.show()
```
